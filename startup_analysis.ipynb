```python
df['deal_flag'] = df['Deal_Status'].astype(str).str.lower().map({
    'yes': 1,
    'y': 1,
    'true': 1,
    'deal': 1,
    'no': 0,
    'n': 0,
    'false': 0,
    'no deal': 0
}).fillna(0).astype(int)
```

```python
df.columns.tolist()
```

```python
df['deal_flag'] = df['Got Deal'].astype(str).str.lower().map({
    'yes': 1,
    'y': 1,
    'true': 1,
    '1': 1,
    'no': 0,
    'n': 0,
    'false': 0,
    '0': 0
}).fillna(0).astype(int)

df['deal_flag'].value_counts()
```

```python
def clean_money(x):
    if pd.isna(x): return None
    return float(str(x).replace(',', '').replace('$','').strip())

df['ask_amount_clean'] = df['Original Ask Amount'].apply(clean_money)
df['deal_amount_clean'] = df['Total Deal Amount'].apply(clean_money)
```

```python
industry_stats = df.groupby('Industry')['deal_flag'].agg(['count', 'sum'])
industry_stats['deal_rate'] = industry_stats['sum'] / industry_stats['count']
industry_stats.sort_values('deal_rate', ascending=False).head(15)
```

```python
import matplotlib.pyplot as plt

industry_stats = industry_stats.sort_values('count', ascending=False).head(15)
industry_stats['deal_rate'].plot(kind='bar', figsize=(10,5))
plt.title("Deal Rate by Industry")
plt.ylabel("Deal Success Rate")
plt.show()
```
